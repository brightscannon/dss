> # 키워드 경험은행 만들기
- FastCampus의 데이터사이언스스쿨에 커리큘럼을 기대를 가지고 즐겁게 배우고있다.
    - 커리큘럼과는 별도로 팀을 만들어서 FinanceData분석을 위한 공부를 함께 진행중
    - 데이터 핸들링을 즐겁게 진행중
    - 데이터수집팀을 만들어서 많은 데이터를 확보하는 크롤링 공부중
- 일을 쉬는동안 내가 하고싶은 일을 찾기위해 여러가지를 도전해왔다.
    - 아두이노와 회로학습을 통해 빛이 밝은 곳으로 이동하는 로봇 제작
    - 3D프린터 3개 조립 후 많은 출력물 출력
    - 3D객체를 설계하여 메카트로닉스 부품을 직접 만들고 조립
    - ROBOTIS 로봇동아리 참여
    - 로봇의 인공지능 구현에 한계점을 느끼고 딥러닝과 텐서플로우를 공부
    - 로봇동아리에 참여하여 ROS에 대해 배우고 사용해본점
- 일을 쉬면서 자산을 모으기 위해 투자를 하며 공부했다.
    - 주식투자 차트를 보며 기술적 분석
    - 자동거래앱을 만들기 위해 파이썬가이드라인 학습
    - 가상화폐를 접하고 선물시장을 경험해봄으로써 틱단위로 
- 글로벌링커스에서 웹 프로젝트를 진행하면서 PHP와 프론트/백앤드를 빠르게 이해하고 적용할 줄 안다고 칭찬받음
    - 회의때 말단사원임에도 개발자와 상사간의 충돌되는 의견의 중재안을 현실성있게 잘 제안하여 새로운 웹의 구조를 개발하는 파트에 참여하게됨.
    - 입사후 1달만에 회사의 쇼핑웹 개발을 담당함.
    - 디자이너와 개발자 사이의 커뮤니케이션 문제를 자주 해결함.
    - 항상 엑셀로 상황과 변수들을 작성하여 문제발생시 빠르게 해결하려 노력함
- 건설회사에 근무하면서 다양한 사람들과 만나고 그들의 삶을 접할수 있게됨
    - 언성이 높아지는 현장 상황에도 사람사이를 잘 중재하는 현장기사
    - 모든 협력업체 사장님들과 원만하게 지냄
    - 현장에서 필요한 가설계도를 빠르게 제작 가능하여 소장님과 현장근로자들의 총애를 받음
