### append

```python
#사람의 이름과 나이가 들어간 데이터를 만듭니다
import random,string

def get_name():
    names = ["Adam","Alan","Alex","Alvin","Andrew","Antony","Arnold"]
    return random.choice(names)
get_name()
```

```python
def get_age(start = 20, end=40):
    return random.randint(start,end)
get_age()
```

```python
def make_data(rows=10):
    datas=[]
    for _ in range(rows):
        data = {"Age":get_age(), "Name":get_name()}
        datas.append(data)
    return datas
make_data()
```

```python
data1 = make_data()
df1 = pd.DataFrame(data1)
df1
```

```python
data2 = make_data()
df2 = pd.DataFrame(data2)
df2
```

```python
#append
#df1에 df2를 추가

df3 = df1.append(df2)
df3
```

```python
#index값 리셋하기
#drop = 새롭게 생성되는 인덱스 컬럼을 삭제합니다.
#inplace = True 는 수정된 데이터가 해당변수에 바로 적용된다
#inplace = True 를 사용하지 않으면 결과 데이터를 받아서 저장해야된다.

df3.reset_index(drop=True, inplace=True)
df3
```

```python
# append를 할때 인덱스를 리셋
df3 = df1.append(df2, ignore_index=True)
df3
```

## merge sql(join)
- 아이디 이름 나이 데이터 프레임 생성
- 아이디, 돈 데이터 프레임을 생성


```python
user_df = pd.DataFrame(columns=["UserID", "Name", "Age"])

for idx in range(1,7): #주의점은 현재 데이터보다 많이 넣으면 무한루프임
    name = get_name()
    
    #중복 이름 제거
    while name in list(user_df["Name"]):
        name = get_name()
    
    #데이터 name_df insert
    data = {"Name":name, "UserID":idx, "Age":get_age()}
    user_df.loc[len(user_df)] = data
    
user_df
```

```python
money_df = pd.DataFrame(columns=["ID","Money"])

for idx in range(15):
    money = random.randint(1,20) *1000
    data = {"Money":money, "ID":random.randint(1,8)}
    money_df.loc[len(money_df)] = data
    
#
money_df
```

```python
# merge - user_df, money_df - key:ID, UserID
money_df.merge(user_df, left_on="ID", right_on="UserID")
```

```python
user_df.merge(money_df, left_on="UserID",right_on="ID")
```

```python
user_df.rename(columns={"UserID":"ID"},inplace=True)
user_df
```

```python
result_df = pd.merge(money_df, user_df)
result_df
```

```python
#사람별로 돈을 얼마나 벌었는지 확인해보자. 

money_list = result_df.groupby("Name").sum()["Money"].reset_index() 
  #sum을 하면 인덱스가 없어지므로 인덱스를 붙이는 차원에서 reset_index삽입
money_list

# Money데이터를 내림차순으로 정렬하고 index를 리셋한다.
money_list = money_list.sort_values(['Money'], ascending=False).reset_index(drop=True)
money_list
```

```python
#merge - outer
#fillna- NaN을 특정데이터로 채워줍니다.
result = pd.merge(user_df, money_list, how='outer').fillna(value=0)

result = result.sort_values(by=['Money'], ascending=False).reset_index(drop=True)
result

```

```python
#change data type 특정 컬럼의 데이터 타입 변경
#float 데이터인 Money 컬럼을 int로 형변환
result["Money"]=result["Money"].astype("int")
result
```
