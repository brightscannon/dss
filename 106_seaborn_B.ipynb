# Seaborn / pandas direct plotting
- marplotlib을 기반으로 하는 더 고급스러운 인터페이스를 제공하기 위해 개발된 python시각화 라이브러리

```python
%matplotlib inline
%config InlineBackend.figure_formats = {'png','retina'}
```

```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def sinplot():
    x = np.linspace(0,10,101)
    for i in range(6):
        plt.plot(x, np.sin(i+x))
sinplot()
```

```python
sns.set()
sinplot()
```

```python
sns.set_style("whitegrid")
sinplot()
```

```python
sns.despine()
sinplot()
```

### set context
- 그래프의 굵기
- paper, notebook, talk, poster

```python
sns.set_context("paper")
sinplot()
```

```python
sns.set_context("notebook")
sinplot()
```

```python
sns.set_context("talk")
sinplot()
```

```python
sns.set_context("poster")
sinplot()
```

### color palettes
- 색상 테마설정이 가능

```python
cp = sns.color_palette()
sns.palplot(cp)
```

```python
p = sns.color_palette("hls", 8)
sns.palplot(p)
```

```python
sns.set_palette("hls",8)
sinplot()
```

```python
# color 테마 리스트는 에러를 발생시켰을때 나오는 정보에 다양하게 들어있으니 확인해보자.
colors = ["muted","hls","RdBu","CMRmap"]
for color in colors:
    p = sns.color_palette(color, 8)
    sns.palplot(p)
```

### Sample data
- https://github.com/mwaskom/seaborn-data

```python
iris = sns.load_dataset("iris")
iris.tail()
```

```python
# histogram
sns.rugplot(iris.petal_length.values)
```

```python
sns.kdeplot(iris.petal_length.values)
```

```python
sns.distplot(iris.petal_length.values, rug=True) # kde를 없애고 싶다면 kde=False 속성을 넣어주자
```

### Joint plot
- 두개의 플롯을 비교해주는 플롯

```python
sns.jointplot(x="sepal_length", y="sepal_width",data=iris)
```

### pair plot

```python
sns.pairplot(iris, hue="species")
plt.show()
```

### heatmap

```python
flight = sns.load_dataset("flights")
flight.tail()
```

```python
flight_data = flight.pivot("month","year","passengers")
flight_data
```

```python
sns.heatmap(flight_data, cmap="YlGnBu", annot=True, fmt="d") # annotation=값 보여주기, formatting=숫자값 포맷 변경
```
