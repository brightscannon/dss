## Thread
- 프로그램이 동작될때 실행시간이 오래 걸린다면 두가지 문제가 있기 때문
- I/O바운드 : 파일을 읽고 쓰거나, 네트워크 시간등의 문제
- cpu바운드 : 계산량이 많은경우 --> 코드의 효율화나 알고리즘을 통해서 해결해야 함(리스트 축약 등)
- I/O바운드의 문제를 Thread를 통해서 어느정도 해결할 수 있음.


- 동기 sync
    - 작업이 끝나야 다음 작업이 수행됨
- 비동기 async
    - 작업이 독립적으로 수행됨
    

```python
import threading
```

```python
ls1, ls2 = [],[]
```

```python
def th_func1(number=1000):
    for idx in range(number):
        ls1.append(idx)
        if idx%100 == 0:
            print("ls1 :",idx)

def th_func2(number=1000):
    for idx in range(number):
        ls2.append(idx)
        if idx%100 == 0:
            print("ls2 :",idx)
```

```python
th1 = threading.Thread(target=th_func1, args=(12000,))
th2 = threading.Thread(target=th_func2, args=(10000,))
th1.start()
th2.start()
```
