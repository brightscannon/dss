# 1. CRM과 카뱅

- 고객상호관리 (customer relationship management)
    - 문자나 등급제 프로그램 등의 활동을 CRM활동이라고 부름
    - 여기서 가장 중요한것이 데이터 분석임
    - 금융에서 데이터 분석이라 하면 일단 CRM과 신용평가 등이 있겠다.
    
- 데이터분석팀 in kakaobank
    - 채널/CRM
        - 데이터 수집 기획, 고객분석 마트설계
        - 고객 행동분석 리포트, 고객활동 예측 모형 개발
        - 개인화 마케팅/서비스 기획& 구현/실행
    - 빅데이터
        - 빅데이터 서비스 개발/운영(CAS.CMS)
        - 데이터 분석 과제 수행(FDS, CSS, OCR등)
        .
        .
        .
        
    - CRM의 미션(kakao에서는?)
        - 고객의 가치를 아는것, 고객의 가치를 증대시키는것
        - 고객 가치평가모델 개발, CRM프로그램 기획/운영

- CRM 워크플로우
    - 데이터기획 -> 데이터수집 -> 데이터분석 -> 분석리포트
    - .........................└-> 분석마트설계 -> 분석마트구축 -> 고객행동모형 개발 -> (개인화서비스 기획 -> 개인화 서비스 구현)[서비스기획팀]
    - ........................................................................................└-> 개인화 마케팅 기획 -> 개인화 마케팅 실행
    

- 데이터 분석 단계(카뱅에서)
    1. 업무영역별 driving table 선정
    2. 분석필요 input정보 추가
    3. pivot, summaty 정보 추가
    4. 업무영역별 분석 table 통합
    - 등등....
    
- 데이터 분석 단계가 가장 시간이 걸린다. ohe를 한다던가 구간화를 한다던지, 평균치를 재산정한다던지 변수 다루는 시간이 가장 오래 걸린다.

- 모델 퍼포먼스를 높이는 가장 좋은 방법중 하나는 변수를 다시 만지는 방법이 있다. 700개가 만들어져도 유의미한 변수는 10~20개 정도??
