```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
```

### bar chart

```python
#수직그래프 - 에러표시, 투명도조절 등

y = np.random.randint(1,10,size=5)
x = range(5)

error = np.random.randint(1,5, size=5)

print(list(x),y,error)
plt.bar(x,y,yerr=error, alpha = 0.5)
plt.show()
```

```python
#수평그래프 - barh, align

y = np.random.randint(1,10,size=5)
x = range(5)

error = np.random.randint(1,5, size=5)

print(list(x),y,error)
plt.barh(x,y,xerr=error, align='center', alpha = 0.5)
plt.show()
```

```python
#누적그래프 - barh, align

y1 = np.random.randint(5,10,size=5)
y2 = np.random.randint(1,5,size=5)
x = range(5)
print(x,y1,y2)
plt.bar(x,y1, alpha = 0.8)
plt.bar(x,y2, alpha = 0.5)
plt.show()
```

### pie chart

```python
#pie plot - labels, colors
y = np.random.randint(1,10,size=5)
labels = ["A","B","C","D","E"]
colors = ["y","gold","lightyellow","r","coral"]
plt.pie(y, labels=labels, colors = colors, autopct="%1.2f%%")
plt.show()
```

### scatter plot

- numpy random rand(uniform), randn(정규분포), randint(균등분포)

```python
#균등분포
y = np.random.rand(500)
x = np.random.rand(500)
plt.figure(figsize=(8,8))
plt.scatter(x,y)
plt.show()
```

```python
#정규분포
y = np.random.randn(500)
x = np.random.randn(500)
plt.figure(figsize=(8,8))
plt.scatter(x,y)
plt.show()
```

```python
y = np.random.randint(1,100, size = 500)
x = np.random.randint(1,100, size = 500)
plt.figure(figsize=(8,8))
plt.scatter(x,y)
plt.show()
```

```python
#원의 크기로 데이터의 크기를 설정 (4차원 데이터)
y1= np.random.randint(1,10,size=25)
y2= np.random.randint(1,10,size=25)

colors = np.random.rand(25)
area = np.pi * np.arange(10,50,2) ** 2  #원의 넓이 계산
# print(y1,y2)
# print(colors, area)

plt.figure(figsize=(5,5))

plt.scatter(y1,y2, s = area, c = colors, alpha = 0.5, marker="*")
plt.show()
```

### histogram

```python
#균등분포 histogram - 쪼개기(bins)

x= np.random.rand(500)
plt.hist(x, bins = 50)
plt.show()
```

```python
#정규분포 histogram

x= np.random.randn(100000)
plt.hist(x, bins = 250)
plt.show()
```
