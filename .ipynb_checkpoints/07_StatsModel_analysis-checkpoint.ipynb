# 회귀분석용 가상데이터 생성방법

https://datascienceschool.net/view-notebook/0224ae9672eb4ab789c395a2ca2335e9/
(입력인수및 출력인수 설명은 링크 참조)

- `X, y, w = make_regression(..., coef=True, ...)`

```python
# 생성해보기 - 독립변수 1개
from sklearn.datasets import make_regression

X, y, w = make_regression(n_samples=10, n_features=1, bias=0, noise=0,
                          coef=True, random_state=0)

X,y,w
```

```python
plt.scatter(X, y, s=100)
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
# noise 인수를 증가시키면  Var[ϵ] 가 증가하고 bias 인수를 증가시키면 y 절편  b 가 증가한다
X, y, w = make_regression(n_samples=40, n_features=1, bias=100, noise=10,
                          coef=True, random_state=0)
plt.scatter(X, y)
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
#연습문제1-1

def make_regression2(n_samples=100, bias=0.0, noise=0.0, random_state=None):
    np.random.seed(random_state)
    X = np.random.randn(n_samples)
#     np.random.seed(None)
    w = np.random.rand() 
    e = np.random.randn(n_samples) * noise/100
    y = w*X + bias + e
    
    return X,y,w


X, y, w = make_regression2(n_samples=50, bias=100, noise=5, random_state=0)

plt.scatter(X, y, s=100)
plt.xlabel("x")
plt.ylabel("y")
plt.show()

```

```python
#독립변수가 2개인 경우!------------------------------------------------

X, y, w = make_regression(n_samples=300, n_features=2, noise=10, 
                          coef=True, random_state=0)

plt.scatter(X[:,0], X[:,1], c=y, s=100, cmap=mpl.cm.bone)
plt.xlabel("x1")
plt.ylabel("x2")
plt.axis("equal")
plt.show()
```

```python
X, y, w = make_regression(n_samples=300, n_features=2, n_informative=1, 
                          noise=0, coef=True, random_state=0)

plt.scatter(X[:,0], X[:,1], c=y, s=100, cmap=mpl.cm.bone)
plt.xlabel("x1")
plt.ylabel("x2")
plt.axis("equal")
plt.show()
```
