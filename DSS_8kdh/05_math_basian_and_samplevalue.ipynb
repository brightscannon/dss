```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
#주피터노트북에서 라인 아래에 보여지도록 설정

%config InlineBackend.figure_formats = {'png','retina'} 
#png 이미지 압축기술설정, retina 그래프 해상도 2배 높히기
```

```python
#1. NumPy를 사용하여 20개의 숫자를 무작위로 생성한 후
#   히스토그램을 그리고 샘플 평균을 구한다.

r = np.random.randint(1,20,size=100)
print(r)
plt.xlim(0,20)
plt.ylim()
plt.hist(r,bins=30)
plt.show()

np.mean(r)
```

```python
#2. 1번과 같이 20개의 숫자 생성 및 샘플 평균 계산을
#   10번 반복하여 10개의 샘플 평균을 구한다.

ls = [np.mean(r),] #1번 문제의 평균도 함께 넣어서 계산
for i in range(50): #1번문제에서 1개가 나왔으므로 9번의 추가계산
    a = np.random.randint(1,20,size=100)
    ls.append(np.mean(a))
print(ls)
```

```python
#3. 2번에서 구한 샘플 평균의 히스토그램을 그리고 샘플 평균을 구한다.

plt.xlim(0,20)
plt.ylim()
plt.hist(ls,bins=30,range=(0,20))
plt.show()

np.mean(ls)
```
