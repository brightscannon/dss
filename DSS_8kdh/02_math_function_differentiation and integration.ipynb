```python
import sympy

# Juypter 노트북에서 수학식의 LaTeX 표현을 위해 필요함
sympy.init_printing(use_latex='mathjax')  
```

```python
#변수선언
x = sympy.symbols('x')
x
```

```python
#함수설정
f = x * sympy.exp(x)
f
```

```python
#미분
sympy.diff(f)
```

```python
#식 재정리
sympy.simplify(sympy.diff(f))
```

```python
#편미분하기-----------------------------------------
x, y = sympy.symbols('x y')
f = x ** 2 + x * y + y ** 2
f
```

```python
#심볼불러와서 식에 적용하기-----------------------------
x, mu, sigma = sympy.symbols('x mu sigma')
f = sympy.exp((x - mu) ** 2 / sigma ** 2)
f
```
